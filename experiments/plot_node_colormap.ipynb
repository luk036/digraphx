
# Node Colormap


Draw a graph with matplotlib, color by degree.


```python
import matplotlib.pyplot as plt
import networkx as nx

num_nodes = 24
digraph = nx.cycle_graph(num_nodes)
pos = nx.spring_layout(digraph, iterations=200)
nx.draw(digraph, pos, node_color=range(num_nodes), node_size=800, cmap=plt.cm.Blues)
plt.show()
```
